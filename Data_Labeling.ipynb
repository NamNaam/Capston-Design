```python
import cv2
import cvlib as cv
 
# open video ( )
vidcap = cv2.VideoCapture('C:/Users/erer0/Videos/VideoProc Converter/concentrate/concentrate.mp4') #  

sample_num = 0    
captured_num = 0
    
# loop through frames
while vidcap.isOpened():
    
    # read frame from webcam 
    status, frame = vidcap.read()
    sample_num = sample_num + 1
    
    if not status:
        break
 
    #    
    face, confidence = cv.detect_face(frame)
    
    print(face)
    print(confidence)
 
    # loop through detected faces
    for idx, f in enumerate(face):
        
        (startX, startY) = f[0], f[1]
        (endX, endY) = f[2], f[3]
 
 
        if sample_num % 8  == 0:
            captured_num = captured_num + 1
            face_in_img = frame[startY:endY, startX:endX, :]
            cv2.imwrite('C:/Users/erer0/Desktop/concentrate_dataset/concentrate_video_face/concentrate'+str(captured_num)+'.jpg', face_in_img) #    
            # cv2.imwrite('C:/Users/erer0/Desktop/not_concentrate_dataset/not_concentrate'+str(captured_num)+'.jpg', face_in_img) #    
 
 
    # display output
    cv2.imshow("captured frames", frame)        
    
    # press "Q" to stop
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
# release resources
vidcap.release()
cv2.destroyAllWindows()
```
